## Scatter Plotly Charts
----------------------

```python
import plotly.express as px
data_x = [0,1,2,3,4]
data_y = [0,1,4,9,16]
fig    = px.scatter(x = data_x,y = data_y)
fig.show()
```

```python
import plotly.express as px
iris_data = px.data.iris()
iris_data.head()
```

```python
## find the relation pattern between sepal_width and sepal_length
fig = px.scatter(iris_data,x='sepal_width',y='sepal_length',trendline="ols")
fig.show()
```

```python
import plotly.express as px

df = px.data.tips()
fig = px.scatter(df, x="total_bill", y="tip", trendline="ols")
fig.show()
```

## Bar Plotly Chart

```python
import plotly.express as px

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]

fig = px.bar(x = fruits,y = sales)
fig.show()
```

```python
import plotly.express as px

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]

fig = px.pie(names = fruits,values = sales, title = "Fruits Distribution")
fig.show()
```

```python
import plotly.express as px

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]
fig = px.pie(names = fruits,values = sales, title = "Fruits Distribution")
fig.update_traces(textinfo='label+value+percent')
fig.show()
```

```python
import plotly.express as px
df = px.data.stocks()
df.head()
```

```python
fig = px.line(df,x='date',y='GOOG',title='Google Stock')
fig.show()
```

```python
import numpy  as np
import pandas as pd

df['date'] = pd.to_datetime(df['date'])
# Convert date to numeric format for regression
df['date_ordinal'] = df['date'].map(pd.Timestamp.toordinal)
# Fit a linear trend line
coeffs = np.polyfit(df['date_ordinal'], df['GOOG'], deg=1)
trend = np.poly1d(coeffs)
# Calculate trend values
df['Trend'] = trend(df['date_ordinal'])
# Plot original stock prices
fig = px.line(df, x='date', y='GOOG', title='Google Stock')
# Add trend line on top
fig.add_scatter(x=df['date'], y=df['Trend'], mode='lines', name='Trend Line', line=dict(color='red'))
# Show the figure
fig.show()
```

```python
import numpy  as np
import pandas as pd

df['date'] = pd.to_datetime(df['date'])
# Convert date to numeric format for regression
df['date_ordinal'] = df['date'].map(pd.Timestamp.toordinal)
# Fit a linear trend line
coeffs = np.polyfit(df['date_ordinal'], df['GOOG'], deg=3)
trend = np.poly1d(coeffs)
# Calculate trend values
df['Trend'] = trend(df['date_ordinal'])
# Plot original stock prices
fig = px.line(df, x='date', y='GOOG', title='Google Stock')
# Add trend line on top
fig.add_scatter(x=df['date'], y=df['Trend'], mode='lines', name='Trend Line', line=dict(color='red'))
# Show the figure
fig.show()
```
