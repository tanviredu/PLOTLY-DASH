```python
import plotly.express as px
df = px.data.stocks()
df.head()
```

```python
fig = px.line(df,x='date',y='GOOG',title='Google Stock')
fig.show()
```

```python
import numpy  as np
import pandas as pd
df['date'] = pd.to_datetime(df['date'])
# Convert date to numeric format for regression
df['date_ordinal'] = df['date'].map(pd.Timestamp.toordinal)
# Fit a linear trend line
coeffs = np.polyfit(df['date_ordinal'], df['GOOG'], deg=1)
trend = np.poly1d(coeffs)
# Calculate trend values
df['Trend'] = trend(df['date_ordinal'])
# Plot original stock prices
fig = px.line(df, x='date', y='GOOG', title='Google Stock')
# Add trend line on top
fig.add_scatter(x=df['date'], y=df['Trend'], mode='lines', name='Trend Line', line=dict(color='red'))
# Show the figure
fig.show()
```

# DASH APP 1

```python
import dash
from dash import dcc,html
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
```

```python
df = px.data.stocks()
df.head()
```

```python
app1 = dash.Dash(__name__)
app1.title = "Stock Dashboard"
charts = []
tick = df.columns[1:]
for symbol in tick:
  dfs = df[[symbol]]
  trace = go.Scatter(x=df['date'],y=dfs[symbol],mode="lines",name=symbol)
  charts.append(trace)
```

```python
layout = go.Layout(
    title="Stock Dashboard",
    xaxis = {'title':'Date'},
    yaxis = {'title':'Price'},
    height=600
)
```

```python
app1.layout = html.Div(
[
    html.H1("Stock Price Dashboard",style={"textAlign":'center'}),
    dcc.Graph(
        id='stock_chart',
        figure={
            'data':charts,
            'layout':layout
        }
    )

])
```

```python
if __name__ == "__main__":
  app1.run()
```

# Health Dashboard

```python
import dash
from dash import dcc,html,Input,Output
import plotly.express as px
```

```python
df = px.data.gapminder()
```

```python
df.head()
```

```python
Years = sorted(df['year'].unique())
```

```python
app = dash.Dash(__name__)
app.title = "Global Health and Welth Relation"
```

```python
app.layout = html.Div(
    [
        html.H1("Global Health and Welth Relation",style={'textAlign':'center'}),
        html.Div(
            [
                html.Label("Select Year:",style={'fontWeight':'bold'}),
                dcc.Dropdown(
                    id="year_dropdown",
                    options=[{'label':i,'value':i} for i in Years],
                    value=Years[0]
                )
            ],
            style={
                'width':'30%',
                'margin':'0 auto',
                'padding':'20px'
            }
        ),
        dcc.Graph(id='scatter-plot')
    ]
)
```

```python
@app.callback(
    Output('scatter-plot','figure'),
    Input('year_dropdown','value')
)
def update_graph(selected_year):
  filtered_df = df[df['year'] == selected_year]
  fig = px.scatter(
      filtered_df,
      x = "gdpPercap",
      y = "lifeExp",
      size = "pop",
      color = "continent",
      hover_name="country",
      log_x=True,
      size_max=60,
      title=f"Life Expectency VS GDP Per Capita in {selected_year}"
  )
  return fig

```

```python
if __name__ == "__main__":
  app.run(debug=True)
```

# IRIS Dataset Project

```python
import dash
import plotly.express as px
import pandas         as pd
from dash             import dcc,html,Input,Output
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]
df.head()
```

```python
app4 = dash.Dash(__name__)
app4.title = "Iris Data Visualization"
```

```python
app4.layout = html.Div(
    [
        html.H1("Iris Data Visualization",style={'textAlign':'center'}),
        html.H2("Scatter Plot"),
        html.Div([
            dcc.Dropdown(
                id='scatter-x',
                options=[{'label':col,'value':col} for col in df.columns[:-1]],
                value = df.columns[0],
                placeholder="Select X-axis",
                style={'width':'35%'}

            ),
            dcc.Dropdown(
                id='scatter-y',
                options=[{'label':col,'value':col} for col in df.columns[:-1]],
                value = df.columns[1],
                placeholder="Select Y-axis",
                style={'width':'35%'}

            ),
            dcc.Checklist(
                id='scatter-species',
                options = [{'label':sp,'value':sp} for sp in df['species'].unique()],
                value = df['species'].unique().tolist(),
                inline=True


            )
        ], style={'display':'flex','gap':'30px'}),
        dcc.Graph(id="scatter-plot"),

        html.H2("Histogram"),
        html.Div([
            dcc.Dropdown(
                id='hist-feature',
                options=[{'label':col,'value':col} for col in df.columns[:-1]],
                value = df.columns[0],
                placeholder="Select a Feature",
                style={'width':'35%'}

            )

        ], style={'display':'flex','gap':'30px'}),
        dcc.Graph(id="histogram")
    ]
)



@app4.callback(
    Output('scatter-plot','figure'),
    Input('scatter-x','value'),
    Input('scatter-y','value'),
    Input('scatter-species','value')
)
def update_scatter(x,y,species_list):
  filtered_df = df[df['species'].isin(species_list)]
  return px.scatter(filtered_df,x=x,y=y,color='species',symbol='species')


@app4.callback(
    Output('histogram','figure'),
    Input('hist-feature','value')
)
def update_histogram(feature):
    return px.histogram(df, x=feature, color='species', barmode='overlay', nbins=10)
```

```python
if __name__ == '__main__':
    app4.run()
```

# Power BI Like Project


```python
import dash
import pandas as pd
import plotly.express as px
from dash import dcc,html,Input,Output
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]
df.head()
```

```python
app = dash.Dash(__name__)
app.title = "Power BI Like Dashboard"
```

```python
app.layout = html.Div(
    [
        html.H1("Power BI Like Dashboard",style={'textAlign':'center'}),
        html.H3("Click a species bar below to filter the scatter plot"),
        dcc.Graph(
            id='bar-chart',
            figure = px.histogram(df,x='species',color='species',title="Species Count")
        ),
        dcc.Graph(
            id='scatter-plot'
        )
    ])

@app.callback(
    Output('scatter-plot','figure'),
    Input('bar-chart','clickData')
)
def update_scatter(clickData):
  if clickData is None:
    filtered_df = df
  else:
    selected_species = clickData['points'][0]['x']
    print(selected_species)
    filtered_df = df[df['species'] == selected_species]
    title = f"Filtered: {selected_species}"

    fig = px.scatter(
        filtered_df,
        x='sepal length (cm)',
        y='sepal width (cm)',
        color='species',
        title=title
    )
    return fig



if __name__ == '__main__':
    app.run()
```
